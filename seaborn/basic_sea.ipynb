## two type of plot
* catplot
* relayplot

```python
sns.relplot(x='total_bill',y='tip', data=data)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip', data=data ,hue='time')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip', hue='time', style='smoker', data=data)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip', hue='time', style='smoker', kind='scatter',row='time',data=data)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip', hue='time', style='smoker', kind='scatter',data=data, col='smoker')
plt.show()
```

```python
sns.catplot(y='total_bill', x='day',data=data)
plt.show()
```

```python
sns.catplot(y='total_bill', x='day', kind='box',data=data)
plt.show()
```

```python
sns.catplot(y='total_bill', x='day', kind='bar',data=data)
plt.show()
```

```python
sns.catplot(y='total_bill', x='day', kind='violin',data=data)
plt.show()
```

```python
sns.catplot(y='total_bill', x='day', kind='boxen',data=data)
plt.show()
```
